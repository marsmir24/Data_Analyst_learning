```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from kmodes.kprototypes import KPrototypes
from sklearn.manifold import TSNE
from clustergram import Clustergram
import urbangrammar_graphics as ugg
```

```python
# Загрузка данных
file_path = 'D:/DO_NOT_REMOVE/Data_Analyst_Junior/19/input_data.xlsx'
data = pd.read_excel(file_path)
```

```python
data.head()
```

```python
data = data.drop("contract_num", axis=1)
```

```python
data.info()
```

```python
# Преобразование категориальных признаков в числовые
label_encoders = {}
for column in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le
```

```python
# Стандартизация числовых признаков
scaler = StandardScaler()
numerical_features = data.select_dtypes(include=['int64', 'float64']).columns
data[numerical_features] = scaler.fit_transform(data[numerical_features])
```

```python
data.head()
```

```python
tsne = TSNE(n_components=2)
data_tsne = tsne.fit_transform(data)
```

```python
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], s=10)
```

```python
clustering = AgglomerativeClustering().fit(data)
np.unique(clustering.labels_)
```

```python
clustering = AgglomerativeClustering(linkage='average').fit(data)
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], s=10, c=clustering.labels_)
```

```python
clustering = AgglomerativeClustering(linkage='complete').fit(data)
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], s=10, c=clustering.labels_)
```

```python
clustering = AgglomerativeClustering(linkage='single').fit(data)
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], s=10, c=clustering.labels_)
```

```python
# Определение оптимального количества кластеров методом локтя для KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Метод локтя')
plt.xlabel('Количество кластеров')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2).fit(data)
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], s=10, c=kmeans.labels_)
```

Вывод: следует разбить клиентов на 2-4 кластера

```python
# Определение оптимального количества кластеров методом силуэтного анализа
silhouette_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(data)
    silhouette_avg = silhouette_score(data, cluster_labels)
    silhouette_scores.append(silhouette_avg)

plt.figure(figsize=(10, 5))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Силуэтный анализ')
plt.xlabel('Количество кластеров')
plt.ylabel('Средний силуэтный коэффициент')
plt.show()
```

Вывод: следует разбить клиентов на 2,4 или 7 кластеров.

```python
# Кластеризация методом K-средних
optimal_clusters = 4  # Определено по методу локтя и силуэтному анализу
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
data['KMeans_Cluster'] = kmeans.fit_predict(data)
```

```python
# Визуализация кластеров K-средних
kmeans = KMeans(n_clusters=4).fit(data)
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], s=10, c=kmeans.labels_)
plt.title('Кластеры K-средних')
plt.show()
```

```python
# Иерархическая кластеризация
agg_clustering = AgglomerativeClustering(n_clusters=optimal_clusters)
data['Agg_Cluster'] = agg_clustering.fit_predict(data)
```

```python
# Визуализация кластеров иерархической кластеризации
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], s=10, c=agg_clustering.labels_)
plt.title('Кластеры иерархической кластеризации')
plt.show()
```

```python
# Кластеризация методом K-Prototypes для категориальных и числовых признаков
categorical_features = data.select_dtypes(include=['int']).columns.tolist()
categorical_indices = [data.columns.get_loc(col) for col in categorical_features]

kproto = KPrototypes(n_clusters=optimal_clusters, random_state=42, init='Cao')
clusters = kproto.fit_predict(data, categorical=categorical_indices)
data['KPrototypes_Cluster'] = clusters
```

```python
# Визуализация кластеров K-Prototypes
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], s=10, c=kproto.labels_)
plt.title('Кластеры K-Prototypes')
plt.show()
```

```python
cgram = Clustergram(range(1, 10), method='hierarchical', linkage='ward')
cgram.fit(data)

ax = cgram.plot(
    figsize=(10, 8),
    line_style=dict(color=ugg.COLORS[1]),
    cluster_style={"color": ugg.COLORS[2]},
)
ax.yaxis.grid(False)
sns.despine(offset=10)
ax.set_title('Hierarchical (scikit-learn)')
```

```python
# Выведем распределение данных по кластерам
print(data['KMeans_Cluster'].value_counts())
print(data['Agg_Cluster'].value_counts())
print(data['KPrototypes_Cluster'].value_counts())
```

```python
# Описательная статистика для каждого кластера KMeans
for cluster in range(optimal_clusters):
    cluster_data = data[data['KMeans_Cluster'] == cluster]
    print(f'Кластер {cluster}:')
    print(cluster_data.describe())

# То же для иерархической кластеризации и K-Prototypes
for cluster in range(optimal_clusters):
    cluster_data = data[data['Agg_Cluster'] == cluster]
    print(f'Кластер {cluster}:')
    print(cluster_data.describe())

for cluster in range(optimal_clusters):
    cluster_data = data[data['KPrototypes_Cluster'] == cluster]
    print(f'Кластер {cluster}:')
    print(cluster_data.describe())
```

```python
# Средняя убыточность для каждого кластера KMeans
for cluster in range(optimal_clusters):
    cluster_data = data[data['KMeans_Cluster'] == cluster]
    print(f'Кластер {cluster} - Средняя убыточность:')
    print(cluster_data[['loss_payout_amt', 'loss_payout_amt_usd']].mean())

# То же для иерархической кластеризации и K-Prototypes
for cluster in range(optimal_clusters):
    cluster_data = data[data['Agg_Cluster'] == cluster]
    print(f'Кластер {cluster} - Средняя убыточность:')
    print(cluster_data[['loss_payout_amt', 'loss_payout_amt_usd']].mean())

for cluster in range(optimal_clusters):
    cluster_data = data[data['KPrototypes_Cluster'] == cluster]
    print(f'Кластер {cluster} - Средняя убыточность:')
    print(cluster_data[['loss_payout_amt', 'loss_payout_amt_usd']].mean())
```

Проект гипотезы для A/B тестирования

1. Определение проблемы
Интервал времени доставки слишком широк и не достаточно точен. Из-за этого клиенты не понимают, когда приедет их заказ и недовольны опозданиями.Из-за чего не делают заказ/уходят к конкурентам.

2. Гипотеза
Демонстрация клиентам точного времени доставки, предсказанного новой моделью, вместо интервала времени приведет к повышению коэффициент конверсии на 10%.

3. Целевые метрики
Основная метрика: коэффициент конверсии (процент пользователей, завершивших покупку).
Вторичные метрики: коэффициента повторных покупок, процент отказов, среднее количество заказов в день в расчете на одного клиента

4. Сегментация пользователей
Новые пользователи, пришедшие на сайт.
Постоянные пользователи,которые делали заявку в последние 6 месяцев.

5. Дизайн эксперимента
Размер выборки: 660 пользователей (330 в контрольной группе и 330 в тестовой группе).
Время проведения теста: 2 недели.
Способ разделения пользователей: случайное распределение.

6. План анализа данных
Статистические тесты: t-тест для сравнения коэффициентов конверсии между контрольной и тестовой группами.
Анализ доверительных интервалов для оценки надежности результатов.

7. Критерии успеха
Гипотеза считается подтвержденной, если коэффициент конверсии в тестовой группе будет на 10% выше, чем в контрольной группе, и p-значение будет меньше 0.05.

8. Риски и предположения
Риски: внешние события (например, праздники, акции конкурентов, погодные условия), которые могут повлиять на поведение пользователей.
Предположения: пользователи предпочтут упрощенную форму и будут чаще завершать покупку.

Дизайн A/B-тест

Цель: позитивно повлиять на удовлетворенность клиентов и улучшить бизнес-показатели.

Метрики: Средняя стоимость заказов в день в расчете на одного клиента, среднее количество заказов в день в расчете на одного клиента



Процент разделения аудитории:
50% пользователей увидят старое предсказание времени доставки с интервалом 20–60 минут;
50% пользователей будет доступен прогноз модели.

Было бы хоршо, если бы метрикиами для отслеживания были:

Основные метрики:

Точность времени доставки: разница между предсказанным временем и фактическим временем доставки.
Удовлетворенность клиентов: оценки и отзывы клиентов после доставки.
Конверсия: процент пользователей, которые оформили заказ после просмотра предсказанного времени доставки.

Вторичные метрики:

Число отмененных заказов: количество заказов, отмененных после предсказания времени доставки.
Средний чек: сумма, потраченная пользователями на заказы.
    
Но в Database: skillbox на Adminer нет данных о часах, только о днях. Нет данных о конверсии или страницах/этапах, нет информации об удовлетворенности.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Данные исследования
X = np.array([3, 3.5, 4, 5, 6, 6.5, 7, 8, 8.5, 9])
Y = np.array([17, 22, 14, 17, 15, 12, 11, 9, 8, 6])
Z = np.array([9.1, 5.5, 12.3, 9.2, 14.2, 16.8, 22.0, 18.3, 24.5, 22.7])
```

```python
# Создание DataFrame
data = pd.DataFrame({
    'Дозировка, мг': X,
    'Время до эффета, мин': Y,
    'Продолжительность действия,ч': Z
})
data
```

```python
# Вычисление корреляционной матрицы
correlation_matrix = data.corr()
correlation_matrix
```

```python
# Визуализация корреляционной матрицы
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Корреляционная матрица')
plt.show()
```

Интерпретация корреляций:

Дозировка и Время, до наступления эффекта от лекарства:
    
    Отрицательная корреляция, так как с увеличением дозировки время до начала действия лекарства, вероятно, уменьшается.
Дозировка и Продолжитеность действия: 
    
    Положительная корреляция, так как большая дозировка может увеличивать продолжительность действия лекарства.
Время, до наступления эффекта от лекарства и Продолжитеность действия: 
    
    Отрицательная корреляция, так как меньшее время до начала действия может ассоциироваться с более длительным эффектом лекарства.

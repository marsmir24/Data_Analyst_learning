```python
# импортируем необходимые библиотеки, классы и функции
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
# отключаем предупреждения
import warnings
warnings.filterwarnings('ignore')
```

```python
# настроим удобное отображение графиков и таблиц
%matplotlib inline
plt.rc('figure', figsize=(18, 3))
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pd.options.display.max_rows = 20
```

```python
# загрузим данные
s = pd.read_csv(r'C:\Users\felix\DA_J\Stats\7.4 FedFunds.csv', 
    header=0, 
    index_col=0, 
    parse_dates=True)
s.head()
```

```python
s.shape
```

```python
# оставим данные только до 2008 года
s = s[:'2007']
s.shape
```

```python
# разобьем данные на обучающую и тестовую выборки так,
# чтобы в тесте оказалось 7 наблюдений
X = s.values
train, test = X[0:len(X)-7], X[len(X)-7:]
```

```python
q = [1, 2, 4, 40, 100]
```

```python
for i in q:
    # обучим MA(i)-модель
    print('\nПостроение МА-модели с q = ', i)
    model = ARIMA(train, order=(0, 0, i))
    model_fit = model.fit()
    
    # выведем сводку по модели
    model_fit.summary()
    
    # выведем константу и коэффициенты модели с помощью атрибута .params
    coef = model_fit.params
    print('Константа и коэффициенты:\n\n%s' % coef)
    
    # взглянем, как был получен первый прогноз
    forecast_frst = coef[0] + (model_fit.resid[-1:] * coef[1])[0]
    print('\nПервый прогноз: ', forecast_frst)
    
    # делаем прогнозы на 7 временных шагов вперед
    # для этого в методе .predict() задаем начало и конец периода прогнозирования
    predictions = model_fit.predict(
        start=len(train), end=len(train)+len(test)-1, dynamic=False)
   
    # напечатаем прогнозы на 7 временных шагов вперед
    for n in range(len(predictions)):
        print('спрогнозированное=%f, фактическое=%f' % (predictions[n], test[n]))
    
    # вычислим значение метрики RMSE
    rmse = np.sqrt(
        mean_squared_error(test, predictions))

    print('\nRMSE на тестовой выборке: %.3f' % rmse)
    
    # визуализируем прогнозы
    plt.figure()  # Создание нового графика
    plt.plot(test, label='тестовые данные')
    plt.plot(predictions, color='red', label='прогнозы')
    plt.title(f'Прогнозы для q = {i}')
    plt.legend()
    plt.show()
```

    для модели с q =  1 RMSE на тестовой выборке: 0.905
    для модели с q =  2 RMSE на тестовой выборке: 0.890
    для модели с q =  4 RMSE на тестовой выборке: 0.863
    для модели с q =  40 RMSE на тестовой выборке: 0.600
    для модели с q =  100 RMSE на тестовой выборке: 1.495

Исходя из результатов, можно сделать следующие выводы:

1)Снижение RMSE при увеличении q от 1 до 40:
С увеличением количества предыдущих элементов временного ряда q, ошибка на тестовой выборке (RMSE) уменьшается. Это указывает на то, что включение большего количества предыдущих наблюдений помогает модели лучше учитывать паттерны временного ряда и делать более точные прогнозы. Это особенно заметно в интервале q=1 до q=40.

2)Оптимум при q=40:
Минимальное значение RMSE достигается при q=40. Это говорит о том, что использование 40 предыдущих значений обеспечивает наилучшую точность модели в прогнозировании временного ряда. Для этого набора данных, 40 предыдущих наблюдений лучше всего отражают структуру временного ряда.

3)Резкое ухудшение при q=100:
Увеличение q до 100 приводит к резкому росту RMSE, что указывает на переобучение модели. Когда модель учитывает слишком большое количество предыдущих значений, она может начать "переподгонять" данные, что ухудшает её способность к обобщению и делает прогнозы менее точными. Это типичная проблема для моделей с избыточной сложностью.

Заключение
Оптимальное значение параметра q для данного временного ряда находится около 40. При небольших значениях q модель недостаточно учитывает информацию из прошлого, что приводит к более высоким значениям RMSE. Однако, при слишком большом q модель становится слишком сложной и начинает переобучаться, что приводит к ухудшению её прогностической способности.

Таким образом, выбор правильного значения q критичен для построения эффективной MA-модели, и в данном случае q=40 является наиболее подходящим значением.

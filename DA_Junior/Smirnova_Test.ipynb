```python
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt
```

```python
# Считывание данных из таблицы "Amazon" 
amazon_data = pd.read_excel('C:/Users/felix/Desktop/Тестовое задание_ Аналитик.xlsx', sheet_name='Amazon')
amazon_data.head()
```

```python
amazon_data.shape
```

```python
# Проверка на пропуски и дубликаты
print("Пропуски в данных:")
print(amazon_data.isnull().sum())
print("\nДубликаты:")
print(amazon_data.duplicated().sum())
```

```python
msno.bar(amazon_data)
```

```python
#Удаление дубликатов
amazon_data_clean = amazon_data.drop_duplicates()
```

```python
amazon_data_clean.shape
```

```python
#Группировка по типу устройства
avg_price_per_device_type_group = amazon_data_clean.groupby('Device Type Group', as_index=False) \
                .agg(avg_price=('Price($)', 'mean'))
avg_price_per_device_type_group
```

```python
# Создание столбчатой диаграммы
plt.figure(figsize=(10, 6))
bars = plt.bar(avg_price_per_device_type_group['Device Type Group'], avg_price_per_device_type_group['avg_price'],\
        color=['blue', 'green', 'orange'], alpha=0.4)

# Добавление подписей и заголовков
plt.xlabel('Device Type Group')
plt.ylabel('Average Price ($)')
plt.title('Average Price per Device Type Group')
plt.xticks(rotation=0)

# Добавление подписей значений на график
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='bottom')

# Показать диаграмму
plt.show()
```

```python
tracking_id_data = pd.read_excel('C:/Users/felix/Desktop/Тестовое задание_ Аналитик.xlsx', sheet_name='Tracking ID')
tracking_id_data.head()
```

```python
tracking_id_data.Site.unique()
```

```python
tags_for_WPSite = tracking_id_data[tracking_id_data['Site']=='WPSite']['Тег'].tolist()
tags_for_WPSite
```

```python
amazon_data_clean
```

```python
data_wpsite=  amazon_data_clean[amazon_data_clean['Tracking ID'].isin(tags_for_WPSite)]
data_wpsite.head()
```

```python
total_fees_for_wpsite = data_wpsite.groupby('Tracking ID', as_index=False) \
                .agg(total_fees=('Ad Fees($)', 'sum')).sort_values(by='total_fees', ascending=False)
total_fees_for_wpsite
```

```python
plt.figure(figsize=(10,5))
plt.barh(total_fees_for_wpsite['Tracking ID'], total_fees_for_wpsite['total_fees'])
plt.title('Tracking ID сайта WPSite в порядке убывания дохода по ним (Ad Fees)', fontsize=14)
plt.xlabel('Total ad_fees,($)')
plt.ylabel('Tracking ID')
for i, val in enumerate(total_fees_for_wpsite['total_fees']):
    plt.text(x = val+3, y =i, s=str(val), ha='center')
plt.show()
```

```python
import re

def decode_binary_string(s):
    return ''.join(chr(int(s[i*8:i*8+8],2)) for i in range(len(s)//8))

with open('Data.csv', 'r') as f:
    file = f.read()
    file = re.sub(r' ', '', file)
    print(decode_binary_string(file))
```

```python
with open('Data.csv', 'r') as f:
    file = f.read()
    file = re.sub(r' ', '', file)
    s = decode_binary_string(file)[44:]
    print(s)
    with open('Data_new.csv', 'w') as fn:
        fn.write(s)
```
